```python
import pandas as pd
df = pd.read_csv('Countries.csv')
```

```python
for column in df.columns:
    if df[column].isnull().sum() > 0:  # Check if the column has missing values
        skewness = df[column].skew()

        # If skewness is between -0.5 and 0.5, use the mean
        if -0.5 <= skewness <= 0.5:
            df[column] = df[column].fillna(df[column].mean())
        # If skewness is outside -0.5 to 0.5, use the median
        else:
            df[column] = df[column].fillna(df[column].median())
```

```python
print(df.isnull().sum())
```

```python
# Check for duplicate rows
duplicate_rows = df.duplicated()
print(duplicate_rows.sum())  # This will print the number of duplicate rows
```
